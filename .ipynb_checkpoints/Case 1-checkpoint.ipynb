```python
import pandas as pd
import matplotlib.pyplot as plt
xls= pd.ExcelFile("Data case 1 final 2025 IIALM.xlsx")
print(xls.sheet_names)
```

```python
df = pd.read_excel(xls, sheet_name="Question 2", header=2)
df.columns = df.columns.str.strip()
print(df.head())
```

```python
# Get the cash flows and zero rates (convert them into appropriate units if necessary)
cash_flows = df["Cash flow (x 1,000)"] *1000 # Cash flows in x1000
zero_rates = df["Zero rates"]  # Zero rates

# Time periods (years)
time_periods = df["Maturity (yrs)"]

# Calculate Present Value (PV) of each cash flow
discount_factors = (1 + zero_rates) ** time_periods
pv_cash_flows = cash_flows / discount_factors

# calculate modified duration
mod_duration_terms = (pv_cash_flows / pv_cash_flows.sum()) *  (time_periods/ (1+zero_rates))
modified_duration = mod_duration_terms.sum()
# Calculate the Price of the liability (sum of PVs)
price = pv_cash_flows.sum()

# Calculate the Modified DV01
modified_dv01 = -modified_duration * price / 10000
modified_dv01_print = modified_dv01/1000000 # for notation

# Results
print(f"Modified Duration: {modified_duration:.2f} years")
print(f"Modified DV01: EUR {modified_dv01_print:.2f} million")
print(pv_cash_flows.sum())
```

```python
maturity_buckets = pd.cut(time_periods, bins=[0, 10, 20, 30, 40, 60], right=True,
                          labels=["1-10 years", "11-20 years", "21-30 years", "31-40 years", "41-60 years"])
bucket_contributions = mod_duration_terms.groupby(maturity_buckets).sum()
print(bucket_contributions.sum())
# Plotting the results
bucket_contributions_percentage = bucket_contributions / bucket_contributions.sum() * 100
bucket_contributions.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Contribution of Maturity Buckets to Total Modified Duration')
plt.xlabel('Maturity Bucket')
plt.ylabel('Contribution to Total Modified Duration (Years)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CHECK CONVEXITY
old_pv = pv_cash_flows.sum()
print(f"Old PV: {old_pv:,.2f}")

shocked_zero_rates = zero_rates - 0.008
shocked_discount_factors = (1 + shocked_zero_rates) ** time_periods
shocked_pv_cash_flows = cash_flows / shocked_discount_factors
new_pv = shocked_pv_cash_flows.sum()
print(f"New PV: {new_pv:,.2f}")

price_change = new_pv - old_pv
print(f"Price Change: {price_change:,.2f}")

dv01_impact = modified_dv01 * -80
print(f"DV01 Impact: {dv01_impact:,.2f}")

mod_convexity_terms = (
    pv_cash_flows * (time_periods**2 + time_periods)
) / (
    ((1 + zero_rates) ** 2) * old_pv
)
modified_convexity = mod_convexity_terms.sum()

# is dit wel de basis point change????
mod_convexity_impact = 0.5 * modified_convexity * (-0.008)**2 * old_pv
print(f"Modified Convexity Impact: {mod_convexity_impact:,.2f}")

total_impact = mod_convexity_impact + dv01_impact
print(f"Total Impact (DV01 + Convexity): {total_impact:,.2f}")

print(f"Unexplained Difference: {price_change - abs(total_impact):,.2f}")
```

```python
FR_old = 1.3
asset_value = FR_old * old_pv
print(f"\nAsset Value (from funding ratio): EUR {asset_value:,.2f}")

# --- 1. DV01 of the Cash Position (1-year Zero Coupon Bond) ---
# Assume cash is invested in a 1-year ZCB at the 1-year zero rate.
# Extract the 1-year zero rate from your data (row where Maturity (yrs) == 1)
r1 = zero_rates.iloc[0] 
DF_1yr = 1 / ((1 + r1) ** 1)  # Actual discount factor for 1 year
price_cash = asset_value * DF_1yr
dv01_cash = price_cash * 0.0001  # 1 bp change
print(f"DV01 of Cash (1-year ZCB): EUR {dv01_cash:,.2f}")


# --- 2. DV01 of the 5-Year Swap ---
# Given: 5-year swap notional = EUR 6,000,000,000; swap rate = 2.2990% (swap rate isn't directly needed for DV01)
notional_5yr = 6_000_000_000

# To compute the annuity factor, we need the actual discount factors.
# You already computed: discount_factors = (1+zero_rates)**time_periods, but the actual DF = 1/discount_factors.
DF_actual = 1 / discount_factors
# Add the actual DF to your DataFrame for convenience:
df["DF_actual"] = DF_actual

# Filter for maturities <= 5 years
df_5yr = df[df["Maturity (yrs)"] <= 5]
annuity_5yr = df_5yr["DF_actual"].sum()
dv01_5yr_swap = notional_5yr * annuity_5yr * 0.0001
print(f"DV01 of 5-year Swap: EUR {dv01_5yr_swap:,.2f}")

# --- 3. Combined DV01 of Existing Hedge (Cash + 5-year Swap) ---
dv01_existing = dv01_cash + dv01_5yr_swap
print(f"Combined DV01 (Cash + 5-year Swap): EUR {dv01_existing:,.2f}")

# --- 4. DV01 Shortfall ---
# Your liability DV01 (modified_dv01) is negative by convention; take the absolute value.
dv01_liab_abs = abs(modified_dv01)
dv01_shortfall = dv01_liab_abs - dv01_existing
print(f"DV01 Shortfall to hedge with 30-year Swap: EUR {dv01_shortfall:,.2f}")

# --- 5. 30-Year Swap DV01 and Required Notional ---
# Filter for maturities <= 30 years to compute the annuity factor for the 30-year swap.
df_30yr = df[df["Maturity (yrs)"] <= 30]
annuity_30yr = df_30yr["DF_actual"].sum()
print(f"Annuity Factor for 30-year Swap: {annuity_30yr:.6f}")

# DV01 per EUR notional for the 30-year swap:
dv01_30yr_per_eur = annuity_30yr * 0.0001
print(f"DV01 per EUR Notional for 30-year Swap: {dv01_30yr_per_eur:.8f}")

# Finally, calculate the required notional for the 30-year swap:
notional_30yr = dv01_shortfall / dv01_30yr_per_eur
print(f"Required Notional for 30-year Swap: EUR {notional_30yr:,.2f}")
```

```python
# Example code with higher decimal precision for small DV01 values

swap_rate_5 = 0.02299
swap_rate_30 = 0.02216
years = np.arange(1, 31)
notional_5 = 6000000000

# Calculate the 30-year swap DV01
mod_dv01_30_fixed_terms = (time_periods * (1 + swap_rate_30) / ((1 + zero_rates)**time_periods)) / (10000 * (1 + zero_rates))
mod_dv01_30_fixed = mod_dv01_30_fixed_terms[:30].sum()
mod_dv01_30_floating = 1 / ((1 + zero_rates[0]) * 10000)
mod_dv01_swap_30 = mod_dv01_30_fixed - mod_dv01_30_floating

# Calculate the target DV01 and how much we need to hedge
target_dv01 = 1 * 1.3 * total_assets
want_to_hedge = target_dv01 - total_assets
print(f"Hedge DV01 needed: {want_to_hedge:,.2f}")

# Cash amount
cash = total_assets
print(f"Cash amount: {cash:,.2f}")

# DV01 of the zero-coupon bond (ZCB)
mod_dv01_zcb = cash / (10000 * (1 + zero_rates[0]))
print(f"Modified DV01 for Zero Coupon Bond (ZCB): {mod_dv01_zcb:,.2f}")

# 5-year swap DV01
mod_dv01_5_fixed_terms = (time_periods * (1 + swap_rate_5) / ((1 + zero_rates)**time_periods)) / (10000 * (1 + zero_rates))
mod_dv01_5_fixed = mod_dv01_5_fixed_terms[:5].sum()
print(f"Modified DV01 for 5-year Swap (Fixed Leg): {mod_dv01_5_fixed:,.8f}")

mod_dv01_5_floating = 1 / ((1 + zero_rates[0]) * 10000)
print(f"Modified DV01 for 5-year Swap (Floating Leg): {mod_dv01_5_floating:,.8f}")

mod_dv01_swap_5 = mod_dv01_5_fixed - mod_dv01_5_floating
print(f"Net Modified DV01 for 5-year Swap: {mod_dv01_swap_5:,.8f}")

# DV01 hedge provided by the 5-year swap
dv01_hedge = mod_dv01_swap_5 * notional_5
print(f"DV01 Hedge from 5-year Swap: {dv01_hedge:,.2f}")

# Re-calculate or reuse the 30-year swap DV01
# (We've already done it above, but here it is again for clarity.)
mod_dv01_30_fixed_terms = (time_periods * (1 + swap_rate_30) / ((1 + zero_rates)**time_periods)) / (10000 * (1 + zero_rates))
mod_dv01_30_fixed = mod_dv01_30_fixed_terms[:30].sum()
mod_dv01_30_floating = 1 / ((1 + zero_rates[0]) * 10000)
mod_dv01_swap_30 = mod_dv01_30_fixed - mod_dv01_30_floating
print(f"Net Modified DV01 for 30-year Swap: {mod_dv01_swap_30:,.8f}")

# Calculate the remaining DV01 after the 5-year swap and the ZCB
remaining_dv01 = want_to_hedge - dv01_hedge - mod_dv01_zcb
print(f"Remaining DV01 to hedge with 30-year Swap: {remaining_dv01:,.2f}")

# Notional required for the 30-year swap
notional_30 = remaining_dv01 / mod_dv01_swap_30
print(f"Notional for 30-year Swap: {notional_30:,.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# Liabilities – Maturity Bucket Contributions (from part (a))
# ----------------------------
maturity_buckets = pd.cut(time_periods, bins=[0, 10, 20, 30, 40, 60], right=True,
                          labels=["1-10 years", "11-20 years", "21-30 years", "31-40 years", "41-60 years"])
bucket_contributions_liab = mod_duration_terms.groupby(maturity_buckets).sum()
print("Total Liability Contribution (should equal modified duration):", bucket_contributions_liab.sum())

# ----------------------------
# Hedging Portfolio – Compute Effective Durations from DV01 Terms
# ----------------------------

# For Cash: use a 1-year zero-coupon bond duration
d_cash = 1 / (1 + zero_rates.iloc[0])

# For the 5-year Swap:
# Use the DV01 terms already computed in your code:
dv01_terms_5 = (time_periods * (1 + swap_rate_5) / ((1 + zero_rates)**time_periods)) / (10000 * (1 + zero_rates))
# Restrict to the first 5 years:
duration_5 = np.sum(time_periods[:5] * dv01_terms_5[:5]) / np.sum(dv01_terms_5[:5])
print(f"Effective Modified Duration for 5-year Swap: {duration_5:.4f} years")

# For the 30-year Swap:
dv01_terms_30 = (time_periods * (1 + swap_rate_30) / ((1 + zero_rates)**time_periods)) / (10000 * (1 + zero_rates))
# Restrict to the first 30 years:
duration_30 = np.sum(time_periods[:30] * dv01_terms_30[:30]) / np.sum(dv01_terms_30[:30])
print(f"Effective Modified Duration for 30-year Swap: {duration_30:.4f} years")

# ----------------------------
# Compute Dollar-Duration (DV01 x Duration) Contributions for Each Hedging Instrument
# ----------------------------
# DV01 values (already computed earlier):
DV01_cash = mod_dv01_zcb                  # Cash (from the zero-coupon bond)
DV01_5 = dv01_hedge                       # 5-year swap (already scaled by notional_5)
DV01_30 = notional_30 * mod_dv01_swap_30   # 30-year swap

# Dollar-duration contributions:
contrib_cash = DV01_cash * d_cash
contrib_5 = DV01_5 * duration_5
contrib_30 = DV01_30 * duration_30

# ----------------------------
# Assign Hedging Contributions to Maturity Buckets
# ----------------------------
# For simplicity, we assume:
# - Cash and 5-year swap contributions belong in the "1-10 years" bucket.
# - The 30-year swap contribution belongs in the "21-30 years" bucket.
hedge_bucket_contrib = {
    "1-10 years": contrib_cash + contrib_5,
    "11-20 years": 0,
    "21-30 years": contrib_30,
    "31-40 years": 0,
    "41-60 years": 0
}
buckets = ["1-10 years", "11-20 years", "21-30 years", "31-40 years", "41-60 years"]
hedge_contrib = pd.Series(hedge_bucket_contrib, index=buckets)

# ----------------------------
# Combined Grouped Bar Plot: Liabilities vs. Hedging Portfolio
# ----------------------------
x = np.arange(len(buckets))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
# Liability contributions (from part (a))
liab_vals = bucket_contributions_liab.reindex(buckets, fill_value=0).values
rects1 = ax.bar(x - width/2, liab_vals, width, label='Liabilities', color='skyblue', edgecolor='black')
# Hedging portfolio contributions
hedge_vals = hedge_contrib.reindex(buckets, fill_value=0).values
rects2 = ax.bar(x + width/2, hedge_vals, width, label='Hedging Portfolio', color='lightgreen', edgecolor='black')

ax.set_xlabel('Maturity Bucket')
ax.set_ylabel('Contribution to Modified Duration (Dollar-Duration)')
ax.set_title('Maturity Bucket Contributions to Total Modified Duration')
ax.set_xticks(x)
ax.set_xticklabels(buckets)
ax.legend()

plt.tight_layout()
plt.show()
```
